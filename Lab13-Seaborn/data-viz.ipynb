```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('who_suicide_statistics.csv')
```

```python
df.head(5)
```

```python
df.info()
```

```python
df.describe()
```

```python
sns.set(style="darkgrid")
```

```python
g = sns.jointplot("year", "suicides_no", data=df, kind="reg",
                  xlim=(1979, 2016), ylim=(0, 23000), color="m", height=17)
plt.show(g)
```

```python
sns.boxplot(x="suicides_no", y="year",
            hue="sex", palette=["m", "g"],
            data=df)
sns.despine(offset=10, trim=True)
```

### Story as I see it...
##### When looking at the number of suicides world-wide in the last thirty years I draw a few conclusions.<br> In the first graph, at the top, the line representing the data regression appears to suggest that suicide rates have increased with the years.<br>Looking to the right of the top graph, we can see that there are some massive outliers in this data set. The vast majority of the data shows that the majority of countries have suicide rates under 200 per year for each year.

```python
sns.set(style="whitegrid")
```

```python
f, ax = plt.subplots(figsize=(6.5, 6.5))
sns.despine(f, left=True, bottom=True)
bedrooms = [5, 4, 3, 2, 1]
sns.scatterplot(x="datesold", y="price",
                hue="bedrooms",
                palette="ch:r=-.2,d=.3_r",
                hue_order=bedrooms,
                sizes=(1, 8), linewidth=0,
                data=houses, ax=ax)
plt.show()
```

```python
pokemon = pd.read_csv("Pokemon.csv")
pokemon.head(5)
```

```python
pokemon.describe()
```

```python
sns.relplot(x="Attack", y="Defense", hue="Generation", size="Generation",
            sizes=(40, 400), alpha=.5, palette=["#9b59b6", "#3498db", "#95a5a6", "#e74c3c", "#34495e", "#2ecc71"],
            height=6, data=pokemon)
```

### Story as I see it...
##### When looking at the pokemon data comparing Attack and Defense, there are a few conclusions that I can draw.<br>This is a scatter plot using various sizes and six hues, one for each of the generations of pokemon.<br>This graph illustrates how most of the unique pokemon have come from the sixth (green) and fourth (red) generations.<br>From this visualization we can see that most pokemon have a fairly even distribution around 75 attack and 75 defense. From there, it appears that the higher the attack, means a lower relative defense score.
